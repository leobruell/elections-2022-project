I think we will start the historic data in 2008 (so we will cover Obama, Trump, and Biden eras). Now I want to get all the partisan scores in one dataframe. 

```python
def get_score(year):
    year_data = senate_elections[senate_elections['year'] == year].copy()
    #as the independent senators in the 2010s caucused with the dems, we will count them as democrats: 
    year_data['party_simplified'] = year_data['party_simplified'].str.replace('OTHER','DEMOCRAT')
    year_data['perc'] = year_data['candidatevotes'] / year_data['totalvotes']
    year_data = year_data[year_data['perc'] > .1]
    results_clean = year_data[['state','party_simplified','perc']].pivot_table(index='state', columns=['party_simplified'])
    results_clean = results_clean.reset_index(level=0)
    if len(results_clean.columns) == 4:
        results_clean.columns = ['state','democrat','other','republican']
    if len(results_clean.columns) == 3:
        results_clean.columns = ['state','democrat','republican']
    #Below, we drop CA because the single primary often results in two dems running against each other in the general.
    results_clean = results_clean[results_clean['state'] != 'CALIFORNIA']
    results_clean['partisan_score_{}'.format(year)] = results_clean['democrat'] - results_clean['republican']
    return results_clean[['state','partisan_score_{}'.format(year)]]
```

```python
def get_score_detail(year):
    year_data = senate_elections[senate_elections['year'] == year].copy()
    #as the independent senators in the 2010s caucused with the dems, we will count them as democrats: 
    year_data['party_simplified'] = year_data['party_simplified'].str.replace('OTHER','DEMOCRAT')
    year_data['perc'] = year_data['candidatevotes'] / year_data['totalvotes']
    year_data = year_data[year_data['perc'] > .1]
    results_clean = year_data[['state','party_simplified','perc']].pivot_table(index='state', columns=['party_simplified'])
    results_clean = results_clean.reset_index(level=0)
    if len(results_clean.columns) == 4:
        results_clean.columns = ['state','democrat','other','republican']
    if len(results_clean.columns) == 3:
        results_clean.columns = ['state','democrat','republican']
    #Below, we drop CA because the single primary often results in two dems running against each other in the general.
    results_clean = results_clean[results_clean['state'] != 'CALIFORNIA']
    results_clean['partisan_score_{}'.format(year)] = results_clean['democrat'] - results_clean['republican']
    return results_clean
```

```python
results = get_score(2020)
```

```python
for year in [2018, 2016, 2014, 2012, 2010, 2008]:
    results = results.merge(get_score(year), how='outer', on='state')
```
